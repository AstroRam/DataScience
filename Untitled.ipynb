```python
from astropy.cosmology import Planck18
import numpy
import matplotlib.pyplot as plt

z = numpy.linspace(0.0,8,100)
print(Planck18.lookback_time(z))
plt.plot(z,Planck18.lookback_time(z).value,'-')
plt.show()
print(Planck18.comoving_distance(z))
plt.plot(z,Planck18.comoving_distance(z).value,'-')
plt.show()
```
